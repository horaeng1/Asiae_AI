```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='ticks')
%matplotlib inline
```

```python
data = pd.read_csv("X-train_encoded_3.csv", encoding='cp949')
```

```python
data2 = pd.read_csv("X_train.csv", encoding='cp949')
```

```python
data
```

```python
data2
```

```python
buyer_nm_f = data2[['buyer_nm']]
part_nm_f = data2[['part_nm']]
str_nm_f = data2[['str_nm']]
```

```python
from sklearn.preprocessing import OrdinalEncoder
ordinal_encoder = OrdinalEncoder()
buyer_cat_encoded=ordinal_encoder.fit_transform(buyer_nm_f)
part_cat_encoded=ordinal_encoder.fit_transform(part_nm_f)
str_nm_cat_encoded=ordinal_encoder.fit_transform(str_nm_f)
```

```python
buyer_encoded = pd.DataFrame(buyer_cat_encoded)
part_encoded = pd.DataFrame(part_cat_encoded)
str_nm_encoded = pd.DataFrame(str_nm_cat_encoded)
```

```python
buyer_encoded.columns=["buyer_encoded"]
part_encoded.columns=["part_encoded"]
str_nm_encoded.columns=["str_nm_encoded"]
```

```python
str_nm_encoded
```

```python
data2=pd.concat([data2,buyer_encoded, part_encoded],axis=1)
```

```python
data2=pd.concat([data2,str_nm_encoded],axis=1)
```
